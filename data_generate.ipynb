```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\tzuli\Documents\python\teco\data\train\train_data.csv")
print(df.shape)
df.head(3)
```

```python
keep_col = [col for col in df.columns if ("角度" not in col) and ("平衡轉速" not in col)]
df = df[keep_col]
keep_col
```

```python
new_col = [col.split("_")[0] for col in keep_col]
df.columns = new_col
print(df.shape)
df.head(3)
```

```python
df_new = pd.DataFrame()
g = df.groupby(["工號"])
for group in df["工號"].unique():
    df_group = g.get_group(group)
    df_id = df_group.loc[:, :"工號"]
    df_group1 = df_group.loc[:, "序號":]

    df_group0 = pd.DataFrame(columns = df_group1.columns, index = df_group1.index)
    df_group2 = df_group1.shift(1)
    df_group3 = df_group1.shift(2)
    df_group4 = df_group1.shift(3)
    df_group5 = df_group1.shift(4)
    df_group6 = df_group1.shift(5)


    df1 = pd.concat([df_id, df_group0.add_prefix(0), df_group0.add_prefix(1), df_group0.add_prefix(2), df_group0.add_prefix(3), df_group0.add_prefix(4), df_group1.add_prefix(5)], axis = 1)
    df2 = pd.concat([df_id, df_group0.add_prefix(0), df_group0.add_prefix(1), df_group0.add_prefix(2), df_group0.add_prefix(3), df_group2.add_prefix(4), df_group1.add_prefix(5)], axis = 1)
    df3 = pd.concat([df_id, df_group0.add_prefix(0), df_group0.add_prefix(1), df_group0.add_prefix(2), df_group3.add_prefix(3), df_group2.add_prefix(4), df_group1.add_prefix(5)], axis = 1)
    df4 = pd.concat([df_id, df_group0.add_prefix(0), df_group0.add_prefix(1), df_group4.add_prefix(2), df_group3.add_prefix(3), df_group2.add_prefix(4), df_group1.add_prefix(5)], axis = 1)
    df5 = pd.concat([df_id, df_group6.add_prefix(0), df_group5.add_prefix(1), df_group4.add_prefix(2), df_group3.add_prefix(3), df_group2.add_prefix(4), df_group1.add_prefix(5)], axis = 1)
    df6 = pd.concat([df_id, df_group6.add_prefix(0), df_group5.add_prefix(1), df_group4.add_prefix(2), df_group3.add_prefix(3), df_group2.add_prefix(4), df_group1.add_prefix(5)], axis = 1)


    df_new = pd.concat([df_new, df1, df2, df3, df4, df5, df6], ignore_index = True)
    df_new = df_new.drop_duplicates(keep = "first")

df_new = df_new.fillna(0)
int_col = [col for col in df_new.columns if ("序號" in col)]
df_new[int_col] = df_new[int_col].astype(int)
```

```python
import pandas as pd
import random
df = pd.read_csv(r"C:\Users\tzuli\Documents\python\teco\data\train\train_data.csv")
df = df.dropna().reset_index(drop = True)
print(df.shape)
df.head(3)
```

```python
df_new = pd.DataFrame()
g = df.groupby("工號")
for group in df["工號"].unique():
    df_group0 = g.get_group(group)
    df_group = df_group0.loc[:, "初始_L側角度":]
    df_new = pd.concat([df_new, df_group], ignore_index = True)
    df_diff = df_group.shift(-1) - df_group
    for _ in range(9):
        df_diff1 = df_diff / 10 * random.uniform(1, 9)
        df_group1 = df_group.loc[:, "初始_L側角度":] + df_diff1
        df_group1 = df_group1.dropna()

        df_new = pd.concat([df_new, df_group1], ignore_index = True)
```

```python
angle_col = [col for col in df.columns if ("角度" in col)]
df_new[angle_col] = df_new[angle_col].round().astype(int)
df_new = df_new.round(2)

check_col = [col for col in df.columns if ("初始" in col) or ("最終" in col)]
df_new = df_new[(df_new[check_col] > 0).all(axis = 1)]

df_new = df_new.drop_duplicates().reset_index(drop = True)
df_new
```
